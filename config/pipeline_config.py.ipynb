```sql
%python

# ============================================================
# STUDY NOTE: pipeline_config.py — Project 2
# ------------------------------------------------------------
# Key differences from Project 1:
#   - Source format: JSON (not Parquet) — semi-structured
#   - New path: checkpoints/ — Auto Loader needs this
#   - Event-based schema — nested JSON fields
#   - GitHub Archive as source (real event data)
#
# GitHub Archive URL pattern:
#   https://data.gharchive.org/YYYY-MM-DD-{hour}.json.gz
#   One file per hour, ~50-200MB compressed
#   Contains real GitHub events: push, PR, issues, stars
# ============================================================

BASE_PATH = "/Volumes/workspace/default/github_streaming"

# ── Medallion layer paths ─────────────────────────────────────
BRONZE_PATH   = f"{BASE_PATH}/bronze/events"
SILVER_PATH   = f"{BASE_PATH}/silver/events"
GOLD_PATH     = f"{BASE_PATH}/gold/event_summary"

# ── Auto Loader checkpoint path ───────────────────────────────
# STUDY NOTE:
#   Checkpoints are how Auto Loader tracks which files
#   it has already processed. Without this, every run
#   would reprocess all files from scratch.
#   One checkpoint per stream — never share checkpoints
#   between different streams.
CHECKPOINT_PATH = f"{BASE_PATH}/checkpoints/bronze_to_silver"

# ── Schema inference location ─────────────────────────────────
# STUDY NOTE:
#   Auto Loader infers schema from the first batch of files
#   and saves it here. On subsequent runs it reuses the
#   saved schema instead of re-inferring — much faster.
#   If schema changes, you can reset by deleting this path.
SCHEMA_PATH = f"{BASE_PATH}/schemas/bronze_schema"

# ── Source: GitHub Archive ────────────────────────────────────
# STUDY NOTE:
#   GitHub Archive stores every public GitHub event.
#   Files are per-hour: 2024-01-01-0.json.gz = midnight hour
#   Each file contains thousands of JSON event objects (NDJSON)
#   NDJSON = Newline Delimited JSON — one JSON object per line
#   This is the most common format for event/log data.
GITHUB_ARCHIVE_BASE = "https://data.gharchive.org"

# Hours to simulate landing (0-23)
# We use a few hours to keep file sizes manageable
LANDING_HOURS = ["0", "1", "2", "3"]

PIPELINE = {
    "source_date":  "2024-01-01",
    "write_mode":   "append",
}

# ── Delta table names ─────────────────────────────────────────
SILVER_TABLE = "default.silver_github_events"
GOLD_TABLE   = "default.gold_github_summary"

# ── Paths summary (printed on import) ────────────────────────
print("=" * 50)
print("Project 2: GitHub Streaming Pipeline")
print("=" * 50)
print(f"Bronze     : {BRONZE_PATH}")
print(f"Silver     : {SILVER_PATH}")
print(f"Gold       : {GOLD_PATH}")
print(f"Checkpoint : {CHECKPOINT_PATH}")
print(f"Schema     : {SCHEMA_PATH}")
print(f"Source date: {PIPELINE['source_date']}")
print("=" * 50)
```
